```python
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.utils import shuffle

import matplotlib.pyplot as plt
```

# Read CSV and Setup DataFrame

```python
raw_data_frame = pd.read_csv('./classification.csv')

raw_data_frame_success = raw_data_frame[raw_data_frame['success'] == 1.0]
raw_data_frame_success.reset_index(inplace=True, drop=True)
raw_data_frame_success = raw_data_frame_success[raw_data_frame_success.index <= 100]

raw_data_frame_not_success = raw_data_frame[raw_data_frame['success'] == 0.0]
raw_data_frame_not_success.reset_index(inplace=True, drop=True)
raw_data_frame_not_success = raw_data_frame_not_success[raw_data_frame_not_success.index <= 100]

data_frame_success = pd.DataFrame(data={
    'x1': raw_data_frame_success.age.tolist(),
    'x2': raw_data_frame_success.interest.tolist(),
    'y1': np.array([1 for i in range (0, 101)]),
    'y2': np.array([0 for i in range (0, 101)]),
})

data_frame_not_success = pd.DataFrame(data={
    'x1': raw_data_frame_not_success.age.tolist(),
    'x2': raw_data_frame_not_success.interest.tolist(),
    'y1': np.array([0 for i in range (0, 101)]),
    'y2': np.array([1 for i in range (0, 101)]),
})

data_frame = pd.concat([data_frame_success, data_frame_not_success])
data_frame = shuffle(data_frame)
data_frame.reset_index(inplace=True, drop=True)

print(data_frame)
```

# Visualize the Data

```python
plt.scatter(data_frame_success.x1, data_frame_success.x2)
plt.scatter(data_frame_not_success.x1, data_frame_not_success.x2)
plt.show()
```

# Learning Steps

### 1. Matrix Weight

```python
w11 = np.random.uniform(-0.05, 0.05)
w12 = np.random.uniform(-0.05, 0.05)
w21 = np.random.uniform(-0.05, 0.05) 
w22 = np.random.uniform(-0.05, 0.05)

W = np.array([[w11, w12], [w21, w22]])
```

### 2. Learning Iteration

```python
#sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
learning_rate = 0.15

for index, row in data_frame.iterrows():
    W_lama = W

    input = np.array([[row.x1], [row.x2]])
    output_hidden = np.dot(W_lama, input)
    output_learn = sigmoid(output_hidden)
    output_actual = np.array([[row.y1], [row.y2]])

    error = output_actual - output_learn
    
    delta_W = np.dot(learning_rate * error * output_learn * (1 - output_learn), input.T)

    W_baru = W_lama + delta_W

    W = W_baru

    prediction = np.argmax(output_learn)
    answer = np.argmax(output_actual)

    print(f"index = {index}, prediction = {prediction}, answer = {answer}")
```
